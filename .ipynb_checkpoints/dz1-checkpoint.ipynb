```python
'''
Задание 2
'''
A = np.array([[-35.0,-9.0,6.0],
              [-5,-34.0,1.0],
              [-7.0,3.0,-33.0]])

B = np.array([[31.0],
              [-29.0],
              [182.0]])
```

```python
#Метод простых итераций
a=A.copy()
b=B.copy()

b[0]=b[0]/a[0][0]
b[1]=b[1]/a[1][1]
b[2]=b[2]/a[2][2]


a[0]=a[0]/a[0][0]
a[1]=a[1]/a[1][1]
a[2]=a[2]/a[2][2]

a=a*-1
a[0][0]=0
a[1][1]=0
a[2][2]=0

xyz=b.copy()
print('0:x=',xyz[0],'y=',xyz[1],'z=',xyz[2])
for i in range(10):
    xyz=np.dot(a,xyz)+b
    print(i+1,':x=',xyz[0],'y=',xyz[1],'z=',xyz[2])
```

```python
#Метод Зейделя итераций#посчитать эпсилон и дельта
a=A.copy()
b=B.copy()

b[0]=b[0]/a[0][0]
b[1]=b[1]/a[1][1]
b[2]=b[2]/a[2][2]


a[0]=a[0]/a[0][0]
a[1]=a[1]/a[1][1]
a[2]=a[2]/a[2][2]

a=a*-1
a[0][0]=0
a[1][1]=0
a[2][2]=0
print(a)
print(b)
xyz=b.copy()
print('0:x=',xyz[0],'y=',xyz[1],'z=',xyz[2])
for i in range(5):
    xyz[0]=np.dot(a,xyz)[0]+b[0]
    xyz[1]=np.dot(a,xyz)[1]+b[1]
    xyz[2]=np.dot(a,xyz)[2]+b[2]
    print(i+1,':x=',xyz[0],'y=',xyz[1],'z=',xyz[2])
```

```python
# Степенной метод вычисления спектрального радиуса

a=A.copy()
w=np.array([1.0,1.0,1.0])
w=w.T
w=w/sqrt(np.dot(w.T,w))
print(0,',w=',w)
for i in range(50):
    v=np.dot(a,w)
    p=np.dot(v.T,w)
    w=v/sqrt(np.dot(v.T,v))
    print(i+1,":v=",v,',w=',w,',p=',p)
```

```python
# Метод вращений Якоби для симметричной матрицы
a=A.copy()
H_list=[]
V=np.eye(3)
for i in range(6):
    print ("Матрица A(",i,')')
    print (a)

    #поиск наибольшего числа
    k,m=0,1
    if fabs(a[0][2])>fabs(a[k][m]):
        k,m=0,2
    if fabs(a[1][2])>fabs(a[k][m]):
        k,m=1,2
    phi=1/2*atan(2*a[k][m]/(a[k][k]-a[m][m]))
    print(i,":",phi)
    H=np.eye(3)
    H[k][k]=cos(phi)
    H[m][m]=cos(phi)        
    H[k][m]=-sin(phi)    
    H[m][k]=sin(phi)
    H_list.append(H)
    a=np.dot(np.dot(H.T,a),H)
for i in range (len(H_list)):
        print ("Матрица V(",i,')')
        print (V)
        V=np.dot(V,H_list[i])
print ("Матрица A")
print (a)
print ("Матирица V")
print (V)
```
